```python
# ======================================================
# ‚úÖ Edge Detection (Custom Image + Auto Save & Download)
# ======================================================

# Step 1Ô∏è‚É£: Import necessary librariesa
import cv2
import matplotlib.pyplot as plt
from google.colab import files
import numpy as np
import zipfile
import os
```

```python
# Step 2Ô∏è‚É£: Upload an image
print("üì§ Please upload your image (JPG/PNG):")
uploaded = files.upload()
```

```python
# Step 3Ô∏è‚É£: Read the uploaded image (as grayscale)
image_path = next(iter(uploaded))
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

if image is None:
    raise ValueError("‚ùå Failed to load image. Please upload a valid image file.")
```

```python
# Step 4Ô∏è‚É£: Apply Gaussian Blur to reduce noise
blurred_image = cv2.GaussianBlur(image, (5, 5), 1.4)
```

```python
# Step 5Ô∏è‚É£: Apply Canny Edge Detection
edges = cv2.Canny(blurred_image, 50, 150)
```

```python
# Step 6Ô∏è‚É£: Display results
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Edge Detected Image (Canny)')
plt.imshow(edges, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Step 7Ô∏è‚É£: Save both images
output_dir = "edge_detection_outputs"
os.makedirs(output_dir, exist_ok=True)

original_path = os.path.join(output_dir, "original_image.png")
edges_path = os.path.join(output_dir, "edge_detected_image.png")

cv2.imwrite(original_path, image)
cv2.imwrite(edges_path, edges)
```

```python
# Step 8Ô∏è‚É£: Zip and download
zip_filename = "edge_detection_results.zip"
with zipfile.ZipFile(zip_filename, "w") as zipf:
    zipf.write(original_path)
    zipf.write(edges_path)

files.download(zip_filename)
print(f"‚úÖ All images saved and downloaded as: {zip_filename}")
```
