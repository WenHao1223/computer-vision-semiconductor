```python
# 2Ô∏è‚É£ Import libraries
import cv2
import matplotlib.pyplot as plt
from google.colab import files
```

```python
# 3Ô∏è‚É£ Upload an image (e.g., wafer_sample.png)
print("üì§ Please upload your image (e.g., wafer_sample.png):")
uploaded = files.upload()
```

```python
# Get uploaded filename
image_path = list(uploaded.keys())[0]
print(f"‚úÖ Uploaded: {image_path}")
```

```python
# 4Ô∏è‚É£ Load and preprocess image
img = cv2.imread(image_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# 5Ô∏è‚É£ Define ROI (adjust x, y, w, h as needed)
x, y, w, h = 0, 140, 200, 200
roi = img_rgb[y:y+h-80, x:x+w-80]
```

```python
# 6Ô∏è‚É£ Plot
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.title("Original (RGB)")
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title("Grayscale")
plt.imshow(img_gray, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title("ROI Focus Area")
plt.imshow(roi)
plt.axis('off')

plt.show()
```

```python
output_filename = "roi.png"
plt.imsave(output_filename, roi)
print(f" Roi image saved as: {output_filename}")
```
