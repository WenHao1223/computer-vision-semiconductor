```python
# ==========================================================
# ✅ Image Enhancement Demo + Auto Save & Download (Colab)
# ==========================================================

# 1️⃣ Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from google.colab import files
import zipfile
import os
```

```python
# 2️⃣ Create output folder
output_dir = "enhanced_outputs"
os.makedirs(output_dir, exist_ok=True)
```

```python
# 3️⃣ Create or upload image (here we create a sample image)
image = np.zeros((300, 500, 3), dtype=np.uint8)
cv2.putText(image, "Sample Image", (50, 150), cv2.FONT_HERSHEY_SIMPLEX, 1.5, (255, 255, 255), 2)

print("Original Image:")
cv2_imshow(image)
```

```python
# 4️⃣ Brightness Adjustment
brightness_factor = 50
bright_image = cv2.add(image, brightness_factor)
print("\nBrightened Image:")
cv2_imshow(bright_image)
```

```python
# 5️⃣ Contrast Enhancement (alpha-beta correction)
alpha = 1.5  # contrast
beta = 20    # brightness
contrast_image = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)
print("\nContrast Enhanced Image:")
cv2_imshow(contrast_image)
```

```python
# 6️⃣ Noise Reduction (Gaussian Blur)
kernel_size = (5, 5)
blurred_image = cv2.GaussianBlur(image, kernel_size, 0)
print("\nBlurred Image (Noise Reduced):")
cv2_imshow(blurred_image)
```

```python
# 7️⃣ Combined Enhancement (Brightness + Contrast)
combined_enhancement = cv2.convertScaleAbs(bright_image, alpha=1.2, beta=10)
print("\nCombined Enhancement:")
cv2_imshow(combined_enhancement)
```

```python
# 8️⃣ Display all with Matplotlib
plt.figure(figsize=(12, 8))

titles = [
    "Original Image",
    "Brightened Image",
    "Contrast Enhanced Image",
    "Blurred Image (Noise Reduced)",
    "Combined Enhancement"
]
images = [image, bright_image, contrast_image, blurred_image, combined_enhancement]

for i, (title, img) in enumerate(zip(titles, images)):
    plt.subplot(2, 3, i + 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# 9️⃣ Save all images
filenames = []
for title, img in zip(titles, images):
    filename = f"{output_dir}/{title.replace(' ', '_').lower()}.png"
    cv2.imwrite(filename, img)
    filenames.append(filename)
```

```python
# 🔟 Zip and download all images
zip_filename = "enhanced_results.zip"
with zipfile.ZipFile(zip_filename, "w") as zipf:
    for file in filenames:
        zipf.write(file)
```

```python
files.download(zip_filename)
print(f"✅ All enhanced images saved and downloaded as: {zip_filename}")
```

```python
# ==========================================================
# ✅ Image Enhancement (Upload Custom Image + Auto Save)
# ==========================================================

# 1️⃣ Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from google.colab import files
import zipfile
import os

# 2️⃣ Upload a custom image
print("📤 Please upload your image (JPG/PNG):")
uploaded = files.upload()

# Get uploaded image path
image_path = list(uploaded.keys())[0]
print(f"✅ Uploaded: {image_path}")

# 3️⃣ Read the uploaded image
image = cv2.imread(image_path)

if image is None:
    raise ValueError("❌ Failed to read the uploaded image. Please try again with a valid file.")

print("Original Image:")
cv2_imshow(image)

# 4️⃣ Create output folder
output_dir = "enhanced_outputs"
os.makedirs(output_dir, exist_ok=True)

# 5️⃣ Brightness Adjustment
brightness_factor = 50
bright_image = cv2.add(image, brightness_factor)
print("\nBrightened Image:")
cv2_imshow(bright_image)

# 6️⃣ Contrast Enhancement (alpha-beta correction)
alpha = 1.5  # Contrast control (1.0-3.0)
beta = 20    # Brightness control (0-100)
contrast_image = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)
print("\nContrast Enhanced Image:")
cv2_imshow(contrast_image)

# 7️⃣ Noise Reduction (Gaussian Blur)
kernel_size = (5, 5)
blurred_image = cv2.GaussianBlur(image, kernel_size, 0)
print("\nBlurred Image (Noise Reduced):")
cv2_imshow(blurred_image)

# 8️⃣ Combined Enhancement (Brightness + Contrast)
combined_enhancement = cv2.convertScaleAbs(bright_image, alpha=1.2, beta=10)
print("\nCombined Enhancement:")
cv2_imshow(combined_enhancement)

# 9️⃣ Display all using Matplotlib
plt.figure(figsize=(12, 8))

titles = [
    "Original Image",
    "Brightened Image",
    "Contrast Enhanced Image",
    "Blurred Image (Noise Reduced)",
    "Combined Enhancement"
]
images = [image, bright_image, contrast_image, blurred_image, combined_enhancement]

for i, (title, img) in enumerate(zip(titles, images)):
    plt.subplot(2, 3, i + 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')

plt.tight_layout()
plt.show()

# 🔟 Save all images
filenames = []
for title, img in zip(titles, images):
    filename = f"{output_dir}/{title.replace(' ', '_').lower()}.png"
    cv2.imwrite(filename, img)
    filenames.append(filename)

# 1️⃣1️⃣ Zip and download all images
zip_filename = f"enhanced_results_{image_path.split(".")[0]}.zip"
with zipfile.ZipFile(zip_filename, "w") as zipf:
    for file in filenames:
        zipf.write(file)

files.download(zip_filename)
print(f"✅ All enhanced images saved and downloaded as: {zip_filename}")
```
