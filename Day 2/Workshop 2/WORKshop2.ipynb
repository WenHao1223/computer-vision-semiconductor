```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as transforms
import segmentation_models_pytorch as smp
from tqdm import tqdm
import albumentations as A
from albumentations.pytorch import ToTensorV2


# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Using device: {device}')


# ==================== CONFIGURATION ====================
class Config:
    # Paths
    TRAIN_IMG_DIR = '/content/drive/MyDrive/Data/train_images'
    TRAIN_MASK_DIR = '/content/drive/MyDrive/Data/train_masks'
    TEST_IMG_DIR = '/content/drive/MyDrive/Data/test_images'
    TEST_MASK_DIR = '/content/drive/MyDrive/Data/test_masks'

    # Model parameters
    ENCODER = 'mit_b2'  # Vision Transformer encoder (MiT-B2 from SegFormer)
    ENCODER_WEIGHTS = 'imagenet'
    CLASSES = 1  # Binary segmentation (adjust for multi-class)
    ACTIVATION = None  # Set to None, we'll apply sigmoid manually

    # Training parameters
    BATCH_SIZE = 4
    LEARNING_RATE = 1e-4
    EPOCHS = 200
    IMAGE_SIZE = 256

    # Binary threshold for mask prediction
    THRESHOLD = 0.5  # Adjust this value if needed (0.3-0.6 typical range)

    # Model save path
    MODEL_PATH = '/content/best_vit_segmentation.pth'


config = Config()


# ==================== DATASET CLASS ====================
class SegmentationDataset(Dataset):
    def __init__(self, image_dir, mask_dir, transform=None):
        self.image_dir = image_dir
        self.mask_dir = mask_dir
        self.transform = transform

        # Get all image files
        self.images = sorted([f for f in os.listdir(image_dir)
                             if f.endswith(('.jpg', '.png', '.jpeg'))])

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        # Load image
        img_name = self.images[idx]
        img_path = os.path.join(self.image_dir, img_name)
        image = np.array(Image.open(img_path).convert('RGB'))

        # Load mask (same name as image)
        mask_name = img_name
        mask_path = os.path.join(self.mask_dir, mask_name)
        mask = np.array(Image.open(mask_path).convert('L'), dtype=np.float32)
        mask = mask / 255.0  # Normalize to [0, 1]

        # Apply transformations
        if self.transform:
            augmented = self.transform(image=image, mask=mask)
            image = augmented['image']
            mask = augmented['mask']

        mask = mask.unsqueeze(0)  # Add channel dimension
        return image, mask


# ==================== TRANSFORMS ====================
def get_train_transforms(image_size):
    return A.Compose([
        A.Resize(image_size, image_size),
        A.HorizontalFlip(p=0.5),
        A.VerticalFlip(p=0.5),
        A.Rotate(limit=35, p=0.5),
        A.Normalize(
            mean=[0.485, 0.456, 0.406],
            std=[0.229, 0.224, 0.225],
        ),
        ToTensorV2(),
    ])


def get_test_transforms(image_size):
    return A.Compose([
        A.Resize(image_size, image_size),
        A.Normalize(
            mean=[0.485, 0.456, 0.406],
            std=[0.229, 0.224, 0.225],
        ),
        ToTensorV2(),
    ])


# ==================== MODEL CREATION ====================
def create_model(encoder, encoder_weights, classes, activation):
    """Create Vision Transformer segmentation model"""
    model = smp.Unet(
        encoder_name=encoder,
        encoder_weights=encoder_weights,
        in_channels=3,
        classes=classes,
        activation=activation,  # None - we'll apply sigmoid manually
    )
    return model


# ==================== TRAINING FUNCTIONS ====================
def dice_coefficient(pred, target, smooth=1e-6):
    """
    Calculate Dice coefficient on BINARY masks
    pred and target should be binary (0 or 1)
    """
    pred = pred.contiguous().view(-1)
    target = target.contiguous().view(-1)

    intersection = (pred * target).sum()
    dice = (2. * intersection + smooth) / (pred.sum() + target.sum() + smooth)
    return dice


def train_epoch(model, loader, optimizer, criterion, device):
    model.train()
    running_loss = 0.0
    running_dice = 0.0

    pbar = tqdm(loader, desc='Training')
    for images, masks in pbar:
        images = images.to(device)
        masks = masks.to(device)

        # Forward pass (logits output)
        outputs = model(images)
        loss = criterion(outputs, masks)

        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # Calculate metrics on BINARY predictions (after thresholding)
        pred_probs = torch.sigmoid(outputs)
        pred_masks_binary = (pred_probs > config.THRESHOLD).float()  # Binarize predictions

        # Binarize ground truth as well (in case it has float values)
        masks_binary = (masks > 0.5).float()

        dice = dice_coefficient(pred_masks_binary, masks_binary)

        running_loss += loss.item()
        running_dice += dice.item()

        pbar.set_postfix({'loss': loss.item(), 'dice': dice.item()})

    epoch_loss = running_loss / len(loader)
    epoch_dice = running_dice / len(loader)
    return epoch_loss, epoch_dice


def validate_epoch(model, loader, criterion, device):
    model.eval()
    running_loss = 0.0
    running_dice = 0.0

    with torch.no_grad():
        pbar = tqdm(loader, desc='Validation')
        for images, masks in pbar:
            images = images.to(device)
            masks = masks.to(device)

            # Forward pass (logits output)
            outputs = model(images)
            loss = criterion(outputs, masks)

            # Calculate metrics on BINARY predictions (after thresholding)
            pred_probs = torch.sigmoid(outputs)
            pred_masks_binary = (pred_probs > config.THRESHOLD).float()  # Binarize predictions

            # Binarize ground truth as well
            masks_binary = (masks > 0.5).float()

            dice = dice_coefficient(pred_masks_binary, masks_binary)

            running_loss += loss.item()
            running_dice += dice.item()

            pbar.set_postfix({'loss': loss.item(), 'dice': dice.item()})

    epoch_loss = running_loss / len(loader)
    epoch_dice = running_dice / len(loader)
    return epoch_loss, epoch_dice


# ==================== VISUALIZATION ====================
def visualize_predictions(model, dataset, device, num_samples=4):
    """Visualize predictions on test images with BINARY masks"""
    model.eval()

    fig, axes = plt.subplots(num_samples, 4, figsize=(16, num_samples * 4))
    if num_samples == 1:
        axes = axes.reshape(1, -1)

    indices = np.random.choice(len(dataset), min(num_samples, len(dataset)), replace=False)

    with torch.no_grad():
        for idx, sample_idx in enumerate(indices):
            image, mask = dataset[sample_idx]

            # Predict
            image_tensor = image.unsqueeze(0).to(device)
            output = model(image_tensor)

            # Apply sigmoid to get probabilities
            pred_mask_prob = torch.sigmoid(output).cpu().numpy()[0, 0]

            # Apply binary thresholding
            pred_mask_binary = (pred_mask_prob > config.THRESHOLD).astype(np.float32)

            # Calculate dice for this sample
            mask_binary = (mask.numpy()[0] > 0.5).astype(np.float32)
            pred_flat = pred_mask_binary.flatten()
            mask_flat = mask_binary.flatten()
            intersection = (pred_flat * mask_flat).sum()
            dice = (2. * intersection + 1e-6) / (pred_flat.sum() + mask_flat.sum() + 1e-6)

            # Denormalize image for visualization
            img_display = image.cpu().numpy().transpose(1, 2, 0)
            img_display = img_display * np.array([0.229, 0.224, 0.225]) + np.array([0.485, 0.456, 0.406])
            img_display = np.clip(img_display, 0, 1)

            # Plot
            axes[idx, 0].imshow(img_display)
            axes[idx, 0].set_title('Original Image')
            axes[idx, 0].axis('off')

            axes[idx, 1].imshow(mask.cpu().numpy()[0], cmap='gray', vmin=0, vmax=1)
            axes[idx, 1].set_title('Ground Truth')
            axes[idx, 1].axis('off')

            # Show probability map
            axes[idx, 2].imshow(pred_mask_prob, cmap='jet', vmin=0, vmax=1)
            axes[idx, 2].set_title(f'Probability Map')
            axes[idx, 2].axis('off')

            # Show binary mask with dice score
            axes[idx, 3].imshow(pred_mask_binary, cmap='gray', vmin=0, vmax=1)
            axes[idx, 3].set_title(f'Binary Mask (Dice={dice:.4f})')
            axes[idx, 3].axis('off')

    plt.tight_layout()
    plt.show()


def save_predictions(model, dataset, device, output_dir='/content/predictions'):
    """Save all predictions as BINARY masks (white=255 for defects, black=0 for no defect)"""
    os.makedirs(output_dir, exist_ok=True)
    model.eval()

    print(f"\n📊 Statistics:")
    defect_count = 0
    no_defect_count = 0
    total_dice = 0.0

    with torch.no_grad():
        for idx in tqdm(range(len(dataset)), desc='Saving binary predictions'):
            image, mask = dataset[idx]

            # Predict
            image_tensor = image.unsqueeze(0).to(device)
            output = model(image_tensor)

            # Apply sigmoid to get probabilities [0, 1]
            pred_mask_prob = torch.sigmoid(output).cpu().numpy()[0, 0]

            # Convert to BINARY mask: White (255) for defects, Black (0) for no defect
            pred_mask_binary = (pred_mask_prob > config.THRESHOLD).astype(np.uint8) * 255

            # Calculate dice for this image
            mask_binary = (mask.numpy()[0] > 0.5).astype(np.float32)
            pred_binary_float = (pred_mask_prob > config.THRESHOLD).astype(np.float32)

            pred_flat = pred_binary_float.flatten()
            mask_flat = mask_binary.flatten()
            intersection = (pred_flat * mask_flat).sum()
            dice = (2. * intersection + 1e-6) / (pred_flat.sum() + mask_flat.sum() + 1e-6)
            total_dice += dice

            # Count images with defects vs no defects
            if pred_mask_binary.max() > 0:
                defect_count += 1
            else:
                no_defect_count += 1

            # Save as PNG with pure binary values
            img_name = dataset.images[idx]
            save_path = os.path.join(output_dir, img_name)
            Image.fromarray(pred_mask_binary, mode='L').save(save_path)

    avg_dice = total_dice / len(dataset)

    print(f'✅ Binary predictions saved to {output_dir}')
    print(f'   Images with defects: {defect_count}')
    print(f'   Images without defects (pure black): {no_defect_count}')
    print(f'   Average Dice Score: {avg_dice:.4f} ({avg_dice*100:.2f}%)')
    print(f'   White pixels (255) = Defected area')
    print(f'   Black pixels (0) = Normal area / No defect')


# ==================== MAIN TRAINING SCRIPT ====================
def main():
    # Create sample directories (if they don't exist)
    os.makedirs(config.TRAIN_IMG_DIR, exist_ok=True)
    os.makedirs(config.TRAIN_MASK_DIR, exist_ok=True)
    os.makedirs(config.TEST_IMG_DIR, exist_ok=True)
    os.makedirs(config.TEST_MASK_DIR, exist_ok=True)

    print("📁 Place your training images in:", config.TRAIN_IMG_DIR)
    print("📁 Place your training masks in:", config.TRAIN_MASK_DIR)
    print("📁 Place your test images in:", config.TEST_IMG_DIR)
    print("📁 Place your test masks in:", config.TEST_MASK_DIR)
    print(f"\n⚠️  Ensure mask filenames match image filenames!")
    print(f"🎯 Binary threshold set to: {config.THRESHOLD}")
    print(f"📊 Dice coefficient calculated on BINARY masks (white areas only)\n")

    # Check if datasets exist
    if len(os.listdir(config.TRAIN_IMG_DIR)) == 0:
        print("❌ No training images found. Please upload your dataset first.")
        return

    # Create datasets
    train_dataset = SegmentationDataset(
        config.TRAIN_IMG_DIR,
        config.TRAIN_MASK_DIR,
        transform=get_train_transforms(config.IMAGE_SIZE)
    )

    test_dataset = SegmentationDataset(
        config.TEST_IMG_DIR,
        config.TEST_MASK_DIR,
        transform=get_test_transforms(config.IMAGE_SIZE)
    )

    print(f"✅ Train dataset size: {len(train_dataset)}")
    print(f"✅ Test dataset size: {len(test_dataset)}")

    # Create dataloaders
    train_loader = DataLoader(
        train_dataset,
        batch_size=config.BATCH_SIZE,
        shuffle=True,
        num_workers=2,
        pin_memory=True
    )

    test_loader = DataLoader(
        test_dataset,
        batch_size=config.BATCH_SIZE,
        shuffle=False,
        num_workers=2,
        pin_memory=True
    )

    # Create model
    print(f"\n🔧 Creating Vision Transformer model with {config.ENCODER} encoder...")
    model = create_model(
        config.ENCODER,
        config.ENCODER_WEIGHTS,
        config.CLASSES,
        config.ACTIVATION
    )
    model = model.to(device)

    # Loss and optimizer
    # BCEWithLogitsLoss expects logits (no sigmoid applied)
    criterion = nn.BCEWithLogitsLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=config.LEARNING_RATE)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, mode='min', patience=3, factor=0.5
    )

    # Training loop
    print(f"\n🚀 Starting training for {config.EPOCHS} epochs...")
    best_dice = 0.0
    history = {'train_loss': [], 'train_dice': [], 'val_loss': [], 'val_dice': []}

    for epoch in range(config.EPOCHS):
        print(f"\n{'='*50}")
        print(f"Epoch {epoch+1}/{config.EPOCHS}")
        print(f"{'='*50}")

        # Train
        train_loss, train_dice = train_epoch(model, train_loader, optimizer, criterion, device)

        # Validate
        val_loss, val_dice = validate_epoch(model, test_loader, criterion, device)

        # Update scheduler
        scheduler.step(val_loss)

        # Save history
        history['train_loss'].append(train_loss)
        history['train_dice'].append(train_dice)
        history['val_loss'].append(val_loss)
        history['val_dice'].append(val_dice)

        print(f"\nTrain Loss: {train_loss:.4f} | Train Dice: {train_dice:.4f} ({train_dice*100:.2f}%)")
        print(f"Val Loss: {val_loss:.4f} | Val Dice: {val_dice:.4f} ({val_dice*100:.2f}%)")

        # Save best model
        if val_dice > best_dice:
            best_dice = val_dice
            torch.save(model.state_dict(), config.MODEL_PATH)
            print(f"✅ Best model saved with Dice: {best_dice:.4f} ({best_dice*100:.2f}%)")

    # Plot training history
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    ax1.plot(history['train_loss'], label='Train Loss')
    ax1.plot(history['val_loss'], label='Val Loss')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Loss')
    ax1.set_title('Training and Validation Loss')
    ax1.legend()
    ax1.grid(True)

    ax2.plot([d*100 for d in history['train_dice']], label='Train Dice')
    ax2.plot([d*100 for d in history['val_dice']], label='Val Dice')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Dice Coefficient (%)')
    ax2.set_title('Training and Validation Dice Coefficient')
    ax2.legend()
    ax2.grid(True)

    plt.tight_layout()
    plt.show()

    # Load best model for testing
    print(f"\n📊 Loading best model for testing...")
    model.load_state_dict(torch.load(config.MODEL_PATH))

    # Visualize predictions
    print(f"\n🎨 Visualizing BINARY predictions...")
    visualize_predictions(model, test_dataset, device, num_samples=min(4, len(test_dataset)))

    # Save all predictions
    print(f"\n💾 Saving all BINARY predictions...")
    save_predictions(model, test_dataset, device)

    print(f"\n✅ Training complete! Best Dice: {best_dice:.4f} ({best_dice*100:.2f}%)")


# ==================== RUN ====================
if __name__ == '__main__':
    main()
```
